```python
import pandas as pd
import requests

from api import DataApi
from config import DATA

RATE_DB = pd.read_csv(DATA + "ratings.csv", usecols=['userId', 'movieId', 'rating'])

# calculate popularity
def calculate_popularity(rates):
    rate_counts = rates.groupby("movieId").count()['rating'].reset_index()
    rate_average = rates.groupby("movieId").mean()['rating'].reset_index()
    popularity = (pd.merge(rate_counts, rate_average, 'inner', 'movieId')
                    .rename({'rating_x' : 'rate_counts', 'rating_y' : 'rate_average'}, axis=1))
    return popularity

popularity = calculate_popularity(RATE_DB)
```

```python
body = "movies/details"
update_api = DataApi(body)
meta = {}
for index, row in popularity.iterrows():
    movieId = int(row['movieId'])
    data = meta['meta'] = row.loc[['rate_counts', 'rate_average']].to_dict()

    update_api.UPDATE_DATA(movieId, data)

    # print progress
    if (index + 1) % 1000 == 0:
        print(f"({index + 1} / {len(popularity)}) 1000 jobs are finished")
```
